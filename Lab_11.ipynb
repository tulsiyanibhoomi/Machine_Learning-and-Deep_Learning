```python
import numpy as np
```

```python
def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0
```

## OR Gate

```python
x_or = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y_or = np.array([0, 1, 1, 1])
```

### Initialize Weights

```python
weights = np.array([0, 0])
bias = 0
lr = 0.5
```

```python
print('X: ', x_or)
print('Y: ', y_or)

print('\nWeights: ', weights)
print('Bias: ', bias)
print('Learning Rate: ', lr)
```

### Manual Calculation

```python
input_sample = x_or[1]
target = y_or[1]

z = np.dot(input_sample, weights) + bias
print(f'Math: ({input_sample[0]} * {weights[0]}) + ({input_sample[1]} * {weights[1]}) + {bias}')
print(f'Sum (z): {z:.2f}')

prediction = step_function(z)
print('Prediction: ', prediction)

error = target - prediction
print('Error: ', error)
```

### Training Loop

```python
epochs = 5

print('TRAINING')

for j in range(epochs):
    print('\nEPOCH: ',(j + 1))

    for i in range(4):
        x = x_or[i]
        y = y_or[i]

        z = np.dot(x, weights) + bias
        pred = step_function(z)

        error = y - pred

        if error != 0:
            weights = weights + (lr * error * x)
            bias = bias + (lr * error * 1)
            print(f'Incorrect output for {x}! Expected: {y}, Output: {pred}')
            print(f'Updated Weights: {weights}, Bias: {bias}')
        else:
            print(f'Correct for {x}')
```
