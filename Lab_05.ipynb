# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

# Read World bank CSV

```python
df = pd.read_csv('WorldBank.csv')
```

```python
df
```

# Perform conditional selection to find - Population ages 15-64 (% of total population)

```python
row = df[df['Indicator Name'] == 'Population ages 15-64 (% of total population)']
```

```python
row
```

# Divide the data into input and output

```python
x = np.arange(1960,2020).reshape(-1,1)
x
```

```python
y = row.values[0][4:-1:].reshape(-1,1)
y
```

# Plot scatter plot of Population ages 15-64 (% of total population)

```python
plt.scatter(x,y)
```

# Feature Scaling (Mandatory for SVR)**
SVR is highly sensitive to the range of data points. If we don't scale (normalize) the data, the model will fail to find the correct hyperplane.

```python
sc_X = StandardScaler()
sc_Y = StandardScaler()

x_scaled = sc_X.fit_transform(x)
y_scaled = sc_Y.fit_transform(y)

print("Scaling Complete")
```

# Fitting SVR on 3 Different Kernel on dataset

```python
svr = SVR(kernel='linear')
svr.fit(x_scaled, y_scaled.ravel())
y_pred = sc_Y.inverse_transform(svr.predict(x_scaled).reshape(-1,1))
plt.scatter(x,y)
plt.scatter(x,y_pred)
```

```python
svr_rbf = SVR()
svr_rbf.fit(x_scaled, y_scaled.ravel())
y_pred_rbf = sc_Y.inverse_transform(svr_rbf.predict(x_scaled).reshape(-1,1))
plt.scatter(x,y)
plt.scatter(x,y_pred_rbf)
```

```python
svr_poly = SVR(kernel='poly')
svr_poly.fit(x_scaled, y_scaled.ravel())
y_pred_poly = sc_Y.inverse_transform(svr_poly.predict(x_scaled).reshape(-1,1))
plt.scatter(x,y)
plt.scatter(x,y_pred_poly)
```
