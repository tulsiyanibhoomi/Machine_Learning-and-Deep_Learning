# Decision Tree Classifier

```python
from sklearn.datasets import load_iris
```

# Import iris data set using sklearn

```python
data = load_iris()
```

```python
data
```

# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

# Divide the data into input and output

```python
x = data['data']
y = data['target']
```

```python
y
```

# Splitting the dataset into the Training set and Test set

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Fitting DecisionTreeClassifier   on dataset

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

# Display Decision Tree

```python
from sklearn.tree import plot_tree
```

```python
# %matplotlib qt
```

```python
plot_tree(model)
```

# Predict the x_test 

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

# Import diabetes.csv dataset 

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

# Check the distribution of the target

```python
df['Outcome'].value_counts()
```

# Replace zeros with NaN for specific columns

```python
cols = ['Insulin', 'SkinThickness', 'BloodPressure', 'BMI', 'Glucose']
```

```python
df[cols] = df[cols].replace(0, np.nan)
```

# Check for missing values

```python
df.isnull().sum()
```

# Fill missing values with median

```python
for i in cols:
    m = df[i].median()
    df[i] = df[i].replace(np.nan, m)
```

```python
df.isnull().sum()
```

# Visualize Distributions

```python
import seaborn as sns
```

```python
sns.pairplot(df, hue='Outcome')
```

# Convert Target data into integer code

```python
df['Outcome'] = df['Outcome'].map({"Diabetic":1, "Non Diabetic":0})
```

```python
df
```

```python
sns.heatmap(df.corr(), annot=True)
```

# Divide the data into input and output

```python
x = df.drop('Outcome', axis=1)
```

```python
y = df['Outcome']
```

# Splitting the dataset into the Training set and Test set

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

# Create Model

```python
dt = DecisionTreeClassifier()
```

# Fitting DecisionTreeClassifier   on dataset

```python
dt.fit(x_train, y_train)
```

# Display Decision Tree

```python
plot_tree(dt)
```

# Predict the x_test 

```python
y_pred = dt.predict(x_test)
```

# Display Training Accuracy

```python
dt.score(x_train, y_train)
```

# Display Test Accuracy

```python
dt.score(x_test, y_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
plt.title(f'Confusion Matrix')
plt.ylabel('Actual Truth')
plt.xlabel('Model Prediction')
plt.show()
```
